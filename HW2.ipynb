3。用 Python 解决本章 2.2.1 中的渡河问题,生成所有的可行方案。

```python
import pandas as pd
import numpy as np
initial_state = [0, 0, 0, 0]  
final_state = pd.Series([1, 1, 1, 1])
final_state.index = ['wolf', 'lamb', 'veg', 'farmer']
state = pd.DataFrame(columns=["wolf", "lamb", "veg", "farmer"])  
state = state.append({'wolf': 0, 'lamb': 0, 'veg': 0, 'farmer': 0}, ignore_index=True)
state_col = state.columns
def farmer(state, num):
    current = state.loc[len(state) - 1][state_col]
    if (current == final_state).all():
        return state
    elif (current['farmer'] == 1 and current['wolf'] == 1 and current['veg'] == 1 and current['lamb'] == 0):
        current['farmer'] = 0  
        state = state.append(current, ignore_index=True)
        current['farmer'] = 1
        current['lamb'] = 1
        state = state.append(current, ignore_index=True)
        state = farmer(state, 0)
        return state
    else:  
        row = current[current == current['farmer']]
        for id in range(num, len(row) - 1):
            current = state.loc[len(state) - 1][state_col]
            current['farmer'] = 1 - current['farmer']
            current[row.index[id]] = 1 - current[row.index[id]]

            if not (((current['wolf'] == current['lamb']) and (current['lamb'] != current['farmer'])) or \
                    ((current['lamb'] == current['veg']) and (current['lamb'] != current['farmer']))):
                state = state.append(current, ignore_index=True)
                if len(state) == len(state.drop_duplicates()):
                    state = farmer(state, 0)
                    return state
                else:
                    state = state.drop_duplicates()
                    if (id != len(row) - 1 - 1):  
                        state = farmer(state, id + 1)
                        return state
                    else:
                        current = state.loc[len(state) - 1][state_col]
                        current['farmer'] = 1 - current['farmer']
                        state = state.append(current, ignore_index=True)
                        state = farmer(state, 0)
                        return state
output = farmer(state, 0)
```

4。用 while 循环实现“笨办法”求解2,使得一旦找到所要的 g 就跳出循环。这样可以减少不必要的循环次数。

```python
c = 2  
tolerance = 1e-6  
g = 1.0  
while abs(g * g - c) > tolerance:
    g = 0.5 * (g + c / g)
print("根号2的近似值:", g)
```

5.在“牛顿法”求解 2的程序中,把 c设为2或 2000 等不同值,观察运行结果

```python
def newton_sqrt_2(initial_guess, tolerance):
    x = initial_guess
    while True:
        new_x = x - (x**2 - 2) / (2 * x)
        if abs(new_x - x) < tolerance:
            return new_x
        x = new_x
c = 2  
tolerance = 1e-6
result = newton_sqrt_2(c, tolerance)
print( result)
```

7.给出c 的三次方根的牛顿迭代式,并使用 Python 进行实现,假设 c=10

```python
def newton_cubic_root(c, initial_guess, tolerance):
    x = initial_guess
    while True:
        new_x = x - (x**3 - c) / (3 * x**2)
        if abs(new_x - x) < tolerance:
            return new_x
        x = new_x
c = 10
initial_guess = 1.0  
tolerance = 1e-6
result = newton_cubic_root(c, initial_guess, tolerance)
print( result)
```

```python
import random
import cmath
def f(x):
    return x**2 + 4*x*cmath.sin(x)
def monte_carlo_integration(f, a, b, num_samples):
    sum = 0
    for _ in range(num_samples):
        x = random.uniform(a, b)
        sum += f(x)
    
    i = (b - a) * (sum / num_samples)
    return i
a = 2
b = 3
num_samples = 1000000 

result = monte_carlo_integration(f, a, b, num_samples)
print(result)
```
